```python
import matplotlib
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline

model_names = ["Euclidean", "Temporal","Compositional","Mean tracker"]
nll = pd.read_csv("model_simulations/nll_matrix.csv").values
posterior = pd.read_csv("model_simulations/posterior_matrix.csv").values



f, ax = plt.subplots(1,1)
plt.title("Posterior probability", size=20)
ax.imshow(posterior, cmap="GnBu")
plt.xticks(np.arange(4), model_names, size=15)
plt.yticks(np.arange(4), model_names, size=15)
# ax.set_xticklabels(model_names, size=15)
# ax.set_yticklabels(model_names, size=15)

plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

for i in range(len(nll)):
    for j in range(len(nll)):
        text = ax.text( i, j, np.around(posterior[i, j], 2),
                       ha="center", va="center", color="k", size=20)

        
plt.gcf().subplots_adjust(bottom=0.35)
plt.savefig("figures/september/nll.eps")


plt.show()
```
